# File & Folder Automation with Python
In this notebook, we use Python to automate repetitive file management tasks such as renaming files, organizing folders, and moving files automatically.

```python
import os # Work with file paths, file names, and directories
import shutil # Move, copy, and organize files easily
from datetime import datetime
```

```python
os.getcwd()
```

```python
folder_path = "C:/Users/..."

os.listdir(folder_path) # List Files in a Folder
```

### Filter Different File Types

```python
### Filter Only PDF Files
pdf_files = []

for file in os.listdir(folder_path):
    if file.lower().endswith(".pdf"):
        pdf_files.append(file)

pdf_files
```

```python
### Filter Only Excel Files
excel_files = []

for file in os.listdir(folder_path):
    if file.lower().endswith((".xlsx", ".xls")):
        excel_files.append(file)

excel_files
```

```python
### Filter Only Word Files
word_files = []

for file in os.listdir(folder_path):
    if file.lower().endswith((".docx", ".doc")):
        word_files.append(file)

word_files
```

```python
# Filter PowerPoint Files

ppt_files = []

for file in os.listdir(folder_path):
    if file.lower().endswith((".pptx", ".ppt")):
        ppt_files.append(file)

ppt_files
```

### Distinguishing Files vs Folders in a Directory

```python
for item in os.listdir(folder_path):
    full_path = os.path.join(folder_path, item)
    print(item, "-> File" if os.path.isfile(full_path) else "Folder")
```

### Rename Files Automatically

```python
# Sort to keep numbering consistent
pdf_files.sort()

for index, filename in enumerate(pdf_files, start=1):
    old_path = os.path.join(folder_path, filename)

    new_name = f"TEST_2026_{index:03}.pdf"
    new_path = os.path.join(folder_path, new_name)

    os.rename(old_path, new_path)

print("PDF files renamed successfully.")
```

#### Organize Files by Date

```python
# Get File Modified Date

source_folder = folder_path
target_folder = ""

os.makedirs(target_folder, exist_ok=True)
```

```python
for file in os.listdir(source_folder):
    file_path = os.path.join(source_folder, file)

    if os.path.isfile(file_path):
        timestamp = os.path.getmtime(file_path)
        date = datetime.fromtimestamp(timestamp)

        year = date.strftime("%Y")
        month = date.strftime("%m")

        destination = os.path.join(target_folder, year, month)
        os.makedirs(destination, exist_ok=True)

        shutil.move(file_path, os.path.join(destination, file))

print("Files organized successfully.")
```
